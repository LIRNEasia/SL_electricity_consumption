```python
import pandas as pd
import os
```

```python
# creating consumption data dictionary

csv_files = ['/Users/chanukaalgama/Desktop/Lacuna improvements/Final Submission/consumption_data/non_smart_meter/monthly_consumption.csv', '/Users/chanukaalgama/Desktop/Lacuna improvements/Final Submission/consumption_data/smart_meter/6hour_interval/smart_6hour_1.csv', '/Users/chanukaalgama/Desktop/Lacuna improvements/Final Submission/consumption_data/smart_meter/15min_interval/smart_15min_1.csv']


data_dict_list = []

for file in csv_files:
    df = pd.read_csv(file, low_memory=False)
    
    if file.endswith('monthly_consumption.csv'):
        data_type = 'non_smart_meter'
    elif file.endswith('smart_6hour_1.csv'):
        data_type = 'smart_meter_6hour'
    elif file.endswith('smart_15min_1.csv'):
        data_type = 'smart_meter_15min'    
    
    for col in df.columns:
        data_dict_list.append({
            "data_format": data_type,
            "column_name": col,
            "data_type": df[col].dtype,
            "missing_values": df[col].isnull().sum()
        })

df_dict = pd.DataFrame(data_dict_list)
```

```python
# creating survey data dictionary

path = '/Users/chanukaalgama/Desktop/Lacuna improvements/Final Submission/survey_data/wave_1'

data_dict_list = []

for root, folder, files in os.walk(path):
    for file in files:

        if file.endswith('.csv'):

            file_path = os.path.join(root, file)

            df = pd.read_csv(file_path, low_memory=False)    
            
            for col in df.columns:
                data_dict_list.append({
                    "data_table": file,
                    "column_name": col,
                    "data_type": df[col].dtype,
                    "missing_values": df[col].isnull().sum()
                })

df_dict = pd.DataFrame(data_dict_list)
```
